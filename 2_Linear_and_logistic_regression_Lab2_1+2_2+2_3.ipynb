```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
urls = 'https://raw.githubusercontent.com/lenhattung/SP26-AI1904-DAP391m/refs/heads/main/lab_part_2/01%20-%20Linear%20and%20logistic%20regression/wines.csv'
wine = pd.read_csv(urls)
url = 'https://raw.githubusercontent.com/lenhattung/SP26-AI1904-DAP391m/refs/heads/main/lab_part_2/01%20-%20Linear%20and%20logistic%20regression/data/titanic.csv'
titanic = pd.read_csv(url)
display(titanic.head())
display(wine)
```

```python
mod_wine = smf.ols(formula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol + type', data = wine)
```

```python
res_wine = mod_wine.fit()
res_wine.summary()
```

```python
mod_titan = smf.glm(formula = 'survived ~ pclass + sex + age + sibsp + parch + fare + embarked',data = titanic,family = sm.families.Binomial())
res_titan = mod_titan.fit()
res_titan.summary()
```

```python
cols_to_scale = ['fixed_acidity', 'volatile_acidity', 'citric_acid',
                 'residual_sugar', 'chlorides', 'free_sulfur_dioxide',
                 'total_sulfur_dioxide', 'density', 'pH', 'sulphates', 'alcohol']

wine_scaled = wine.copy()
wine_scaled[cols_to_scale] = (wine[cols_to_scale] - wine[cols_to_scale].mean()) / wine[cols_to_scale].std()


mod_mnlogit = sm.MNLogit.from_formula(
    formula='quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol + type',
    data=wine_scaled
)


res_mnlogit = mod_mnlogit.fit(method='bfgs', maxiter=1000)

print(res_mnlogit.summary())
```
